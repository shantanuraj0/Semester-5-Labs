### Problem 6

```python
def unique_ele_list(li):
    s=set()
    for ele in li:
        s.add(ele)
    return s
```

```python
li = [1,1,3,2,3,2,3,2,2]
uniq_ele = unique_ele_list(li)
print(uniq_ele)
```

### Problem 7

```python
def first_repeating_ele(li):
    s=set()
    for ele in li:
        if(ele in s):
            return ele
        s.add(ele)
    
```

```python
li = [1,2,3,4,5,1,2]
first_rep_ele = first_repeating_ele(li)
print(first_rep_ele)
```

### Problem 8

```python
def n_to_dict(n):
    d={}
    for i in range(0,n+1):
        li=[i**2,i**3]
        d[i]=li
    return d
```

```python
n=int(input())
d=n_to_dict(n)
print(d)
```

### Problem 10

```python
def gen_squares(n):
    li=[]
    for i in range(0,n+1):
        li.append(i)
    
    li_new = [ele**2 for ele in li]
    return li_new
    
    
```

```python
n=int(input())
li = gen_squares(n)
print(li)
```

### Problem 12

```python
class MyClass():
    def __init__(self,li):
        self.li = li[:]
    
    def apply(self,func_name):
        try:
            result = list(map(func_name,self.li))
            return result
        except TypeError:
             print("data-type not matched!")
            
```

```python
def sq(x): 
     return x**2
```

```python
c1 = MyClass([1,2,3,4])
print(c1.apply(lambda x:x**2))
[1,4,9,16]
c2 = MyClass(['a','b','c'])
c2.apply(sq)
```
