```python
import matplotlib.pyplot as plt

link_id = [1, 2, 3, 4, 5, 6, 7, 8,9]

#802.11a : throughput1
throughput1 = [1.118, 1.817, 1.527, 1.978, 1.678, 0.6109, 0.5769, .510089, .5446 ]

#802.11g : throughput2
throughput2 = [1.104, 1.813, 1.534, 1.966, 1.672, 0.6093, 0.5774, .510099, .5473 ]
plt.plot(link_id, throughput1 ,"b" )
plt.plot(link_id, throughput2 ,"y")

plt.title('Link Throughput Comparasion')
plt.xlabel('link Id')
plt.ylabel('throughput')
plt.show()
```
