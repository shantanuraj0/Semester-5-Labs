**2nd question**

```python
# Randomly splitting the data
import numpy as np
from sklearn.model_selection import train_test_split

data = pd.read_csv("student_scores.csv")
xdata = data.iloc[:,:-1]
ydata = data.iloc[:,-1]
print("x data: ")
print(xdata)
print("\n")
print("y data")
print(ydata)
x_train,x_test,y_train,y_test = train_test_split(xdata,ydata,test_size=0.1,random_state=1) 
```

```python
# printing the splitted data
print("x train data: ")
print(x_train)
print("\n")
print("x test data: ")
print(x_test)
print("\n")
print("y train data: ")
print(y_train)
print("\n")
print("y test data: ")
print(y_test)
```

**3rd** **question**

```python
# calculating mean , median , mode ,standard deviation 

#Over all
print("Overall mean: ")
print(data.mean().mean())
print("\n")

print("Overall median: ")
print(data.median().median())
print("\n")

print("Overall mode: ")
print(data.mode().mode())
print("\n")

print("Overall standard deviation: ")
print(data.std())
print("\n")



# column wise 

print("column wise mean: ")
print(data.mean(axis=0))
print("\n")

print("column wise median: ")
print(data.median(axis=0))
print("\n")
print("column wise mode: ")
print(data.mode(axis=0))
print("\n")
print("column wise standard deviation: ")
print(data.std(axis=0))
print("\n")

# Row wise
print("row wise mean: ")
print(data.mean(axis=1))
print("\n")
print("row wise median: ")
print(data.median(axis=1))
print("\n")
print("row wise mode: ")
print(data.mode(axis=1))
print("\n")
print("row wise standard deviation: ")
print(data.mode(axis=1))
print("\n")
```
