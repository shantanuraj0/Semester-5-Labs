**1st**

```python
import pandas as pd

data = pd.read_csv("A3_Monday.csv",usecols= [1,2,3,4,5,6,7])

shape = data.shape
print("Number of features:", shape[1]-1)
print("Number of patterns:",shape[0])

print("\nRange of output in dataset:")
print("Minimum output:", data["price"].min(), "Maximum output:", data["price"].max())
```

**2nd**

```python
data_frame = pd.DataFrame(data)

for ratio in range(10, 91, 10):
  x = data_frame.sample(frac = ratio/100)
  y = data_frame.drop(x.index)
  print("\n",ratio, "% in x")
  print(x)
  print(y)
```

**3rd**

```python
from scipy import stats
import numpy as np

series = data_frame.stack()


print("MEAN:")
print("\nColumn wise:")
print(data_frame.mean(axis = 0))
print("\nRow wise:")
print(data_frame.mean(axis = 1))
print("\nOverall:")
print(series.mean())

print("\n\nMEDIAN")
print("\nColumn wise:")
print(data_frame.median(axis = 0))
print("\nRow wise:")
print(data_frame.median(axis = 1))
print("\nOverall:")
print(series.median())

print("\n\nMODE")
print("\nColumn wise:")
print(data_frame.mode(axis = 0))
print("\nRow wise:")
print(data_frame.mode(axis = 1))
print("\nOverall:")
print(series.mode())

print("\n\nSTANDARD DEVIATION:")
print("\nColumn wise:")
print(data_frame.std(axis = 0))
print("\nRow wise:")
print(data_frame.std(axis = 1))
print("\nOverall:")
print(series.std())
```

```python
import matplotlib.pyplot as plt

studentdata = pd.read_csv("student_scores.csv")
stud_data_frame = pd.DataFrame(studentdata)

hours = stud_data_frame['Hours'].tolist()
scores = stud_data_frame['Scores'].tolist()

plt.scatter(hours, scores)

plt.xlabel('Hours')
plt.ylabel('Scores')

plt.show()
```
