```python
from datetime import datetime
from dateutil import relativedelta

# taking inputs
year = int(input("Enter the year "))
month = int(input("Enter the month "))
day = int(input("Enter the day "))

# datetime: Its a combination of date and time along with the attributes year, month, day, hour, minute, second, microsecond, and tzinfo. 
dob = datetime(year, month, day)
# Pandas to_datetime() method helps to convert string Date time into Python Date time object
today = pd.to_datetime("now").date()

age = relativedelta.relativedelta(today, dob)

years = age.years
months = age.months
days = age.days

print("Your age is", years, "yrs,", months, "months and", days, "days")
```
